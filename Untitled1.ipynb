```python
# Cell 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for our plots for a professional look
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

print("Libraries imported successfully!")
```

```python
# Cell 2: Load the dataset
# Make sure you have uploaded "IMDb movies.csv" to your Colab environment
file_name = 'imdb.csv'

try:
    df = pd.read_csv(file_name)
    print("Dataset loaded successfully!")
except FileNotFoundError:
    print(f"Error: Could not find the file '{file_name}'.")
    print("Please make sure you have uploaded the file to Google Colab.")

# Display the first 5 rows to understand the data
if 'df' in locals():
    print(df.head())
```

```python
# Cell 3: Initial Data Inspection
if 'df' in locals():
    # Get a quick overview of all columns, their types, and missing values
    print("\n--- Data Info ---")
    df.info()
```

```python
# Cell 4: Identify Missing Values
if 'df' in locals():
    print("\n--- Missing Values (Count) ---")
    # Calculate the percentage of missing values for each column
    missing_percentage = (df.isnull().sum() / len(df)) * 100
    print(missing_percentage.sort_values(ascending=False))
```

```python
# Cell 5: Cleaning Strategy
if 'df' in locals():
    # 1. Drop columns that are not needed for this analysis or have too many missing values
    # (e.g., description, reviews_from_users, reviews_from_critics)
    cols_to_drop = ['imdb_title_id', 'original_title', 'description',
                    'reviews_from_users', 'reviews_from_critics',
                    'metascore', 'budget', 'usa_gross_income', 'worlwide_gross_income']
    df_cleaned = df.drop(columns=cols_to_drop, errors='ignore')

    # 2. Handle missing 'duration' (numerical) - Impute with the median
    # Median is better than mean because it's not affected by outliers
    median_duration = df_cleaned['duration'].median()
    df_cleaned['duration'].fillna(median_duration, inplace=True)

    # 3. Handle missing 'avg_vote' (our target!) - We must drop rows where the rating is missing
    df_cleaned.dropna(subset=['avg_vote'], inplace=True)

    # 4. Handle missing categorical data (e.g., 'director', 'actors') - Fill with 'Unknown'
    df_cleaned['director'].fillna('Unknown', inplace=True)
    df_cleaned['actors'].fillna('Unknown', inplace=True)

    # 5. Handle missing 'genre' - This is crucial
    df_cleaned.dropna(subset=['genre'], inplace=True)

    print("Data cleaning complete.")
    print("\n--- Missing Values After Cleaning ---")
    print(df_cleaned.isnull().sum())
```

```python
# Cell 6: Data Type Conversion (Professional Step)
# 'year' and 'duration' should be numeric, but 'year' might be an object.
if 'df_cleaned' in locals():
    # Convert 'year' to numeric, forcing errors (like 'TV Movie 2019') to NaN
    df_cleaned['year'] = pd.to_numeric(df_cleaned['year'], errors='coerce')

    # Drop any rows where 'year' could not be converted
    df_cleaned.dropna(subset=['year'], inplace=True)

    # Convert 'year' to integer
    df_cleaned['year'] = df_cleaned['year'].astype(int)

    # Convert 'duration' to integer
    df_cleaned['duration'] = df_cleaned['duration'].astype(int)

    print("\n--- Data Types After Conversion ---")
    df_cleaned.info()
```

```python
# Cell 7: Summary Statistics
if 'df_cleaned' in locals():
    # Get statistics for numerical columns
    print("\n--- Summary Statistics (Numerical) ---")
    # We are interested in 'avg_vote', 'duration', 'year', 'votes'
    print(df_cleaned[['avg_vote', 'duration', 'year', 'votes']].describe())

    # Get statistics for categorical columns
    print("\n--- Summary Statistics (Categorical) ---")
    print(df_cleaned[['genre', 'country', 'director']].describe())
```

```python
# Cell 8: Distribution of Ratings (Visualization)
if 'df_cleaned' in locals():
    plt.figure(figsize=(14, 7))
    sns.histplot(df_cleaned['avg_vote'], kde=True, bins=30, color='blue')
    plt.title('Distribution of Movie Ratings (avg_vote)', fontsize=16)
    plt.xlabel('Average Rating', fontsize=12)
    plt.ylabel('Number of Movies', fontsize=12)
    plt.axvline(df_cleaned['avg_vote'].mean(), color='red', linestyle='--', label=f'Mean: {df_cleaned["avg_vote"].mean():.2f}')
    plt.axvline(df_cleaned['avg_vote'].median(), color='green', linestyle='-', label=f'Median: {df_cleaned["avg_vote"].median():.2f}')
    plt.legend()
    plt.show()
```

```python
# Cell 9: Average Rating by Year
if 'df_cleaned' in locals():
    # Group data by year and calculate the mean rating
    # We filter for years with a good number of movies (e.g., after 1960) to avoid noise
    yearly_ratings = df_cleaned[df_cleaned['year'] > 1960].groupby('year')['avg_vote'].mean().reset_index()

    plt.figure(figsize=(15, 7))
    sns.lineplot(data=yearly_ratings, x='year', y='avg_vote')
    sns.regplot(data=yearly_ratings, x='year', y='avg_vote', scatter=False, color='red', ci=None, label='Trend Line') # Adds a trend line
    plt.title('Average Movie Rating Over Time (Post-1960)', fontsize=16)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Average Rating (avg_vote)', fontsize=12)
    plt.legend()
    plt.show()
```

```python
# Cell 10: Preprocessing for Genre Analysis
if 'df_cleaned' in locals():
    # 1. Create a new dataframe by splitting the 'genre' column
    # 'Action, Comedy' -> 'Action' (row 1), 'Comedy' (row 1)
    df_genre = df_cleaned.copy()
    df_genre['genre_split'] = df_genre['genre'].str.split(', ')

    # 2. 'Explode' the dataframe to create a new row for each genre
    df_genre_exploded = df_genre.explode('genre_split')

    print(f"Original shape: {df_genre.shape}")
    print(f"Exploded shape: {df_genre_exploded.shape}")
    print("\n--- Top 10 Genres ---")
    print(df_genre_exploded['genre_split'].value_counts().head(10))
```

```python
# Cell 11: Visualization - Top 15 Genres by Average Rating
if 'df_genre_exploded' in locals():
    # Calculate average rating for each genre
    genre_ratings = df_genre_exploded.groupby('genre_split')['avg_vote'].mean().reset_index()

    # Filter genres that have at least 500 movies (to remove niche genres)
    genre_counts = df_genre_exploded['genre_split'].value_counts()
    common_genres = genre_counts[genre_counts > 500].index

    # Filter our ratings data to only include these common genres
    top_genre_ratings = genre_ratings[genre_ratings['genre_split'].isin(common_genres)]
    top_genre_ratings = top_genre_ratings.sort_values(by='avg_vote', ascending=False)

    # Plot
    plt.figure(figsize=(16, 8))
    sns.barplot(data=top_genre_ratings, x='avg_vote', y='genre_split', palette='magma')
    plt.title('Average Rating for Common Genres (min. 500 movies)', fontsize=16)
    plt.xlabel('Average Rating', fontsize=12)
    plt.ylabel('Genre', fontsize=12)
    plt.show()
```

```python
# Cell 12: Correlation Heatmap
if 'df_cleaned' in locals():
    # Let's see what features correlate with 'avg_vote'
    corr_matrix = df_cleaned[['avg_vote', 'duration', 'year', 'votes']].corr()

    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Matrix of Numerical Features', fontsize=16)
    plt.show()
```
